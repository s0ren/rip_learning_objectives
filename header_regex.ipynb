# Find og fjern headere fra sider

Øverst på alle sider er der en slags header med denne (eller lignende tekst):

>    Metalindustriens Uddannelsesudvalg  
    Uddannelsesordning for data- og kommunikationsuddannelsen  
    Uddannelsesordning for 1205 Data- og kommunikationsuddannelsen (version 10)  
    Bekendtgørelse om data- og kommunikationsuddannelsen (01-08-2024)  
    Elevtypesamling: Ungdom og EUV3  
    Fag fælles for hovedforløb  
    Øvrige  

Den øverste del (de første 4 linjer), er headeren. Men årstallet i linje 4 ændres (hvert år smart-ass), og jeg vil også gerne at scriptet kan bruges til andre uddannelser end "data- og kommunikationsuddannelsen" eller mere specifikt, andre end "1205 Data- og kommunikationsuddannelsen (version 10)".

Så jeg skal bruge et regulært udtryk der matcher et mønster lignende dette...

Jeg tænker lige jeg kan checke med regex101 <https://regex101.com/r/IaTAmv/4>

```python
"""
code genereret af regex101.com, fra linket ovenfor
"""

# coding=utf8
# the above tag defines encoding for this document and is for Python 2.x compatibility

import re

regex = (r"Metalindustriens Uddannelsesudvalg\n"
	r"Uddannelsesordning for ([^\(\n)]+)\s*\n"
	r"Uddannelsesordning for ([^\n]+)\n"
	r"Bekendtgørelse om (.*\s)?(\1) \((\d\d-\d\d-\d\d\d\d)\)")

test_str = ("Metalindustriens Uddannelsesudvalg\n"
	"Uddannelsesordning for data- og kommunikationsuddannelsen\n"
	"Uddannelsesordning for 1205 Data- og kommunikationsuddannelsen (version 10)\n"
	"Bekendtgørelse om data- og kommunikationsuddannelsen (01-08-2024)\n"
	"Elevtypesamling: Ungdom og EUV3\n"
	"Fag fælles for hovedforløb\n"
	"Øvrige\n\n"
	"Metalindustriens Uddannelsesudvalg\n"
	"Uddannelsesordning for data- og kommunikationsuddannelsen\n"
	"Uddannelsesordning for 1205 Data- og kommunikationsuddannelsen (version 8)\n"
	"Bekendtgørelse om data- og kommunikationsuddannelsen (01-08-2024)\n"
	"Elevtypesamling: Ungdom og EUV3\n"
	"Fag fælles for hovedforløb\n"
	"Øvrige\n\n"
	"Metalindustriens Uddannelsesudvalg\n"
	"Uddannelsesordning for cykelsmedeuddannelsen\n"
	"Uddannelsesordning for 007 Cykel- og pedalkraft (samt elcykel) (version 101)\n"
	"Bekendtgørelse om cykelsmedeuddannelsen (01-08-2027)\n"
	"Elevtypesamling: Ungdom og EUV3\n"
	"Fag fælles for hovedforløb\n"
	"Øvrige\n\n"
	"--- Herunder eksempel på en anden uddannelsesordning, fra Metalindustrien ---\n\n"
	"Metalindustriens Uddannelsesudvalg\n"
	"Uddannelsesordning for Cykel- og motorcykelmekaniker\n"
	"Uddannelsesordning for 1260 Cykel-og motorcykelmekaniker (version 8)\n"
	"Bekendtgørelse om erhvervsuddannelsen til cykel- og motorcykelmekaniker (01-08-2024)\n"
	"Elevtypesamling: Ungdom og EUV3\n"
	"Fag fælles for hovedforløb\n"
	"Øvrige\n")

matches = re.finditer(regex, test_str, re.MULTILINE | re.IGNORECASE)

for matchNum, match in enumerate(matches, start=1):
    
    print ("Match {matchNum} was found at {start}-{end}: {match}".format(matchNum = matchNum, start = match.start(), end = match.end(), match = match.group()))
    
    for groupNum in range(0, len(match.groups())):
        groupNum = groupNum + 1
        
        print ("Group {groupNum} found at {start}-{end}: {group}".format(groupNum = groupNum, start = match.start(groupNum), end = match.end(groupNum), group = match.group(groupNum)))

# Note: for Python 2.7 compatibility, use ur"" to prefix the regex and u"" to prefix the test string and substitution.
```

```python
import re

sampe_header_24 = r'''
Metalindustriens Uddannelsesudvalg
Uddannelsesordning for data- og kommunikationsuddannelsen
Uddannelsesordning for 1205 Data- og kommunikationsuddannelsen (version 10)
Bekendtgørelse om data- og kommunikationsuddannelsen (01-08-2024)
Elevtypesamling: Ungdom og EUV3
Fag fælles for hovedforløb
Øvrige
'''
# print(sampe_header_24)

header_regex_pattern  = r'''
Metalindustriens Uddannelsesudvalg
Uddannelsesordning for ([^\(\n)]+)
Uddannelsesordning for ([^\n]+)
Bekendtgørelse om (.*\s)?(\1) \((\d\d-\d\d-\d\d\d\d)\)
'''
print(header_regex_pattern)

m = re.search(header_regex_pattern, sampe_header_24, re.MULTILINE | re.IGNORECASE)
if m:
    print(f"fandt header fra pos {m.start()} til {m.end()} = \n[{sampe_header_24[m.start() : m.end()]}],\n")
    print(f"with subgroups: {m.groups()} ")
else:
    print(m)
```
